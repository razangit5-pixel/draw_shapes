# Import Modules

```python
import numpy as np
import cv2
```

# Read the Image 

```python
img = cv2.imread(r'draw_shapes\avacado.png',1)
```

# Create black window with zeros function

```python
img =np.zeros((512,512,4),np.uint8)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img 
```

# Draw a line

```python
#pt1 = loc of the line, pt2 = end point, color = color of the line, thickness = width
img = cv2.line(img, (0,0), (255,255), (147, 96, 44), 10) # 44, 96, 147 #draw line
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Draw arrow

```python
img = cv2.arrowedLine(img, (0,255), (255,255), (255, 0, 0), 10) #draw arrow
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Draw rectangle

```python
img = cv2.rectangle(img, (384, 0), (510, 128), (0, 0, 255), 10) #draw rectangle
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.rectangle(img, (384, 0), (510, 128), (0, 0, 255), -1) #draw rectangle
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Draw a circle 

```python
img = cv2.circle(img, (447, 63), 63, (0, 255, 0), -1) #draw circle
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Write a text 

```python
font = cv2.FONT_HERSHEY_SIMPLEX #type of font
img = cv2.putText(img, 'OpenCv', (10, 500), font, 4, (0, 255, 255), 10, cv2.LINE_AA)# write a text
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX #type of font
img = cv2.putText(img, 'OpenCv', (10, 500), font, 4, (0, 255, 255), 10, cv2.LINE_8)# write a text
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32) #make array to draw it by polylines
pts = pts.reshape((-1,1,2))#reshape it
# True to close the polyline
img = cv2.polylines(img,[pts],True,(0,255,255)) #draw polyline
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
